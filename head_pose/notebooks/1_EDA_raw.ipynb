```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
import gc
import os
import json
```

```python
DATA_PATH = '../data/JET_data/processed_data/'

patients = [i for i in os.listdir(DATA_PATH) if i.find('.')==-1]

print(f'Total number of patients: {len(patients)}')

records = []


for patient in patients:
    
    record = [DATA_PATH+patient+'/'+i[:-4] for i in os.listdir(DATA_PATH+patient) if i.find('.npy')!=-1]
    
    records+=record
    
print(f'Total number of records {len(records)}')
```

# Explore target 

```python
labels = []
ages = []

for record in records:
    
    ecg = np.load(record+'.npy')
    ann = json.load(open(record+'.json'))
    
    label = ann['label'][1]
        
    labels.append(int(label))
    ages.append(ann['age'])

labels = np.array(labels)
ages = np.array(list(set(ages)))

n_labels = labels.shape[0]
print(f'Class balance: \n Other ECG: {labels[labels==0].shape[0]} \n JET: {labels[labels==1].shape[0]} ')
```

# Distribution of the age

```python
sns.displot(ages)
```

# Visualize records

```python
n = 10000
records_sub = records[n:n+10]

for record in records_sub:
    
    patient = record.split('/')[-2]
    
    ecg = np.load(record+'.npy')
    ann = json.load(open(record+'.json'))
    

    fig = plt.figure(figsize=(30,10))
    plt.plot(ecg[:5000])
    plt.title(f"Label: {ann['label'][0]}, Patient: {patient}")
    
    
```
